### My code is a function that generates a Fibonacci sequence given the first two integers and a sequence length.

```python
def fibonacci(a,b,L):
    output = [a, b]
    counter = 2
    while counter < L:
        output.append(output[-1] + output[-2])
        counter += 1
    return output

print(fibonacci(1,1,15))
```
